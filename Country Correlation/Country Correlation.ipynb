### 3: HDI Intra-National and International-- By Region

#### 3a: Comparisions Between HDIs of Different Regions of a Given Country, Over Time
__Models Created:__ <br>
> __hdi_sub_national_trends__: For a name of a country, list the region HDI's & total HDI of that given country. <br>
__hdi_sub_national_trends_table__: For a name of a country, list the region & total HDIs, ordered by difference in HDI over time for a given region. <br>
hdi_sub_national_trends_graph <br>

While HDI is commonly evaluated over a nationwide standpoint, HDI could also be calculated from a regional perspective. Certainly, there are particular regions within countries which experience a higher HDI growth than other regions; for instance, the region containing the capital of a given country typically experiences greater HDI growth than those of rural regions.  If this instance were the case, a country's national HDI may not accurately represent the whole of an entire country, as the HDI may skewed lower or higher.  <br>

As an example, take the Philippines.  While the National Capital Region, where the capital Manila resides, experiences a high degree of development, the Southern rural regions are relatively underdeveloped:

Therefore, this model aims to highlight the discrepencies of development within the sub-national partitions of each country.

The following code will take an input, which will be the string of the name of the country, and then will output HDI, by year, organized by region.

```python
def hdi_sub_national_trends(country_name):
    # Retriving sub-national HDIs of selected country, removing years with no value.
    df_country = df[df['Country'].str.match(country_name)].dropna(axis=1)
    
    # df_country.columns.values[6] retrives earliest year recorded.  df_country.columns.values[-1] retrives latest.
    df_new = df_country[df_country.columns.values[-1]]-df_country[df_country.columns.values[6]]
    
    # Taking the difference between oldest and newest HDI.
    df_new = pd.DataFrame({'Difference': df_new})
    
    # Joining new data.
    df_joined = df_country.join(df_new)
    
    return df_joined
```

```python
hdi_sub_national_trends('Belarus')
```

The following code sorts by Difference, and returns Region, earliest and latest HDIs by year, and Difference.

```python
def hdi_sub_national_trends_table(country_name):
    return hdi_sub_national_trends(country_name).sort_values(by="Difference", ascending=False).iloc[:, [5,6,-2,-1]]
```

```python
hdi_sub_national_trends_table('Belarus')
```

The following code line graphs HDI yearly for each region of the specified country.  This graphical representation hopefully provides a clearer perspective of the HDI differences between regions.

```python
def hdi_sub_national_trends_graph(country_name):
    # Retrieve HDI for each year.
    df_cleaned = hdi_sub_national_trends(country_name).drop(
        ['Country','ISO_Code3', 'ISO_Code2', 'Level', 'GDLCODE', 'Difference'], axis=1)
    df_cleaned_years = df_cleaned.drop(['Region'], axis=1)
    
    # Rearrange data into new dataframe for plotting.
    df_new = pd.DataFrame(columns=df_cleaned['Region']) 
    
    new_column = []
    for array in df_cleaned.values:
        new_column = array
        df_new[array[0]] = new_column
    df_new = df_new.drop(index=0, axis=0)
    
    # Create plot.
    
    return df_new.plot(), print('The graph displays the Human Development Index for', country_name, 'under "Total" and its regions over time.')
```

```python
hdi_sub_national_trends_graph('Kuwait')
```

#### 3b. Comparisions Between HDIs of Different Regions of Countries

As shown above, the human development of a country's capital or of the total (average) is not reflective of development of the entire country.  Therefore, a using a singular HDI to represent the human development of the entire country leaves out discrepencies between regions.  So, perhaps, comparing the regions' HDIs rather than comparing that of the countries' HDIs might reflect a more accurate situation.

This following module is simple; it will take one region of a country, and then will output its HDI rank of all the regions of all the countries, and similar HDIs of countries, and regions of countries.  This module will output results only for a particular year.

The following code returns the names of the regions of each country.

```python
def region_names_of_country(country_name):
    # Use hdi_sub_national_trends to get cleaned dataframe.
    df_result = hdi_sub_national_trends(country_name)
    df_result = df_result.drop([0])
    
    # Retrive region names.
    df_result = df_result['Region']
    return df_result
```

```python
region_names_of_country('Afghanistan')
```
